```python
from pymatgen import MPRester
from pymatgen.electronic_structure.bandstructure import BandStructure
from jupyter_jsmol.pymatgen import quick_view
from pymatgen.electronic_structure.plotter import BSPlotter
from matplotlib import pyplot as plt
```

```python
with MPRester("XS47CSTtG0pOGgu7k5") as m:
    structure = m.get_structure_by_material_id("mp-984")
    bandstructure = m.get_bandstructure_by_material_id("mp-984")
    print(bandstructure)
```

```python
band=bandstructure.get_band_gap()
print(band)
print("la bande d'énergie vaut", band['energy'], "eV")
```

```python
plotband=BSPlotter(bandstructure)
plotband.get_plot()
```

```python
premiere_bande_de_conduction=bandstructure.get_cbm()
derniere_bande_de_valence=bandstructure.get_vbm()
print(premiere_bande_de_conduction)
print(derniere_bande_de_valence)

```
