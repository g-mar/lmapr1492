```python
from pymatgen import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
```

```python
with MPRester("XS47CSTtG0pOGgu7k5") as m:
    structure = m.get_structure_by_material_id("mp-984")
```

```python
calculator= XRDCalculator(1.54060,  symprec=0, debye_waller_factors=None)
pattern=calculator.get_pattern(structure, scaled=True, two_theta_range=(0, 90))
print (pattern)
```

```python
#angles, intensité et indices hkl du premier pic
angle1 = pattern.x[0]
intensite1 = pattern.y[0]
indiceshkl1 = pattern.hkls[0]
print(angle1)
print(intensite1)
print(indiceshkl1)
```

```python
#angles, intensité et indices hkl du second pic
angle2 = pattern.x[1]
intensite2 = pattern.y[1]
indiceshkl2 = pattern.hkls[1]
print(angle2)
print(intensite2)
print(indiceshkl2)
```

```python
#angles, intensité et indices hkl du troisième pic
angle3 = pattern.x[2]
intensite3 = pattern.y[2]
indiceshkl3 = pattern.hkls[2]
print(angle3)
print(intensite3)
print(indiceshkl3)
```

```python
#plot des 3 premiers pics ( 2 teta compris entre 20° et 45°)
calculator.show_plot(structure, two_theta_range=(20,45),annotate_peaks=True,ax=None,with_labels=True)
```
